```python
import matplotlib.pyplot as plt
with open('IPL/deliveries.csv','r+') as f:
    data=f.readlines()
    d=[]
    for line in data:
        d.append(line.strip('\n').split(','))

    dict1={}
    for i in range(1,len(d)):
        dict1[d[i][2]]=0
       
    for i in range(1,len(d)):
        dict1[d[i][2]]+=int(d[i][17]) 


    x=list(dict1.keys())
    y=list(dict1.values())
    plt .plot(x,y)
    plt.xlabel('teams')
    plt.ylabel('total_runs_scored')
    plt.title('Ipl Teams & Total runs scored')
    plt.show()
    
 
    
```

```python
import matplotlib.pyplot as plt
with open('IPL/umpires.csv','r+') as f:
    data=f.readlines()
    d=[]
    for line in data:
        d.append(line.strip('\n').split(','))
   
    
    umpairs=[]
    for i in range(1,len(d)):
        umpairs.append(d[i][1].strip())
    

    dt={}
    for i in umpairs:
        if i!='India':
            dt[i]=umpairs.count(i)

    x=list(dt.keys())
    y=list(dt.values())
    plt .plot(x,y)
    plt.xlabel('Countries')
    plt.ylabel('number of umpires')
    plt.title('Foriegn umpais in IPL')
    plt.show()
    
    


    



```

```python
import matplotlib.pyplot as plt
with open('IPL/matches.csv','r+') as f:
    data=f.readlines()
    d=[]
    for line in data:
        d.append(line.strip('\n').split(','))
    seasons=[]
    teams=[]
    for i in range(1,len(d)):
        seasons.append(d[i][1])
        teams.append(d[i][4])
    seasons=sorted(list(set(seasons)))
    teams=sorted(list(set(teams)))
    season_count= [0]*len(seasons)
    plays=[0]*len(teams)
    seasons_data=dict(zip(seasons,season_count))
    teams_plays=dict(zip(teams,plays))

  

    for i in range(1,len(d)):
        teams_plays[d[i][4]]+=1
        seasons_data[d[i][1]]+=1

    plt.bar(seasons_data.keys(),seasons_data.values())
    plt.show()
    plt.bar(teams_plays.keys(),teams_plays.values())
    plt.show()

    
    
```
